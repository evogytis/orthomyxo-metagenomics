```python

def annotate_lineage_discovery(tree,sample_discovery_dates):
    """
    Given a baltic tree and a dict that maps tip names to their discovery date 
    mark all internal branches with the oldest date when a branch was discovered.
    """

    for k in tree.getExternal(): ## iterate over tips
        if 'discovery' in k.traits: ## if tip has been marked already
            del k.traits['discovery'] ## reset trait if this cell was run accidentally before

    for k in sorted(tree.getExternal(),key=lambda w: int(sample_discovery_dates[w.name])): ## iterate over tips in increasing order of discovery
        cur_node=k ## begin at tip

        while cur_node: ## descend to root

            if 'discovery' in cur_node.traits: ## branch already marked before, break out of loop
                break

            cur_node.traits['discovery']=int(sample_discovery_dates[k.name]) ## assign branch tip's discovery date

            cur_node=cur_node.parent ## descend to parent
            
    return tree


def compute_cumulative_discovery(tree,sample_discovery_dates=None):
    """
    Given a baltic tree and a dict (optional) that maps tip names to their discovery date
    returns a list of cumulative diversity contributed by each tip, in their order of discovery.
    If discovery year dict is not provided the tip order is randomised each time.
    """
    ys=[]
    done=set()
    
    if sample_discovery_dates:
        sorted_viruses=sorted(tree.getExternal(),key=lambda k: int(k.traits['discovery'])) ## sort viruses by their discovery order
    else:
        sorted_viruses=np.random.choice(tree.getExternal(),size=len(tree.getExternal()),replace=False) ## resample tips, i.e. will shuffle order
        
    for x in sorted_viruses: ## will compute cumulative diversity in tree after each virus

        cur_node=x ## start at tip
        path_length=0.0 ## reset path length

        while cur_node and cur_node not in done: ## while not at root or at node marked previously
            path_length+=cur_node.length ## add branch length to cumulative
            done.add(cur_node) ## mark branch as done
            cur_node=cur_node.parent ## go to parental branch

        ys.append(sum([k.length for k in done])) ## add sum of all branch lengths encountered in tree so far
        
    return ys


def compute_branch_discovery(tree,sample_discovery_dates):
    """
    Given a baltic tree and a dict that maps tip names to their discovery date
    store the sum of branch lengths contributed by that tip to a dict.
    """
    
    contribs={} ## store contributions for each year

    for L in sorted(set([int(x) for x in sample_discovery_dates.values()])): ## iterate over unique years of discovery
        contribs[L]={}

        for k in tree.getExternal(lambda w: int(w.traits['discovery'])==L): ## iterate over tips discovered same year
            
            contribs[L][k.name]=0 ## set contribution for that year's tip to 0

            cur_node=k ## begin at tip

            while cur_node: ## descend to root

                if cur_node==None or cur_node.traits['discovery']!=L: ## reached either root or a node that's been discovered previously, break out of loop
                    break

                contribs[L][k.name]+=cur_node.length ## add to path length for that tip
                cur_node=cur_node.parent ## descend to parent
                
    return contribs


def diversity_function(x,A,B): ## function to fit
    """
    compute cumulative diversity given discovery index x
    """
    return A*np.log2(1+ np.array(x)/B)


# print(discovery)
#### remember to call the functions
```
