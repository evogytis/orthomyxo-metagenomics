```python
for k in ll.getExternal(): ## iterate over tips
    if 'discovery' in k.traits: ## if tip has been marked already
        del k.traits['discovery'] ## reset trait if this cell was run accidentally before

for k in sorted(ll.getExternal(),key=lambda w: int(discovery[w.name])): ## iterate over tips in increasing order of discovery
    cur_node=k ## begin at tip
    
    while cur_node: ## descend to root
        
        if 'discovery' in cur_node.traits: ## branch already marked before, break out of loop
            break
        
        cur_node.traits['discovery']=int(discovery[k.name]) ## assign branch tip's discovery date
        
        cur_node=cur_node.parent ## descend to parent
```

```python
### this cell occasionally chokes on curve fitting, may need to be rerun until it works
import numpy as np
import scipy.optimize as opt

xs=range(1,len(discovery)+1) ## order of viruses

N=1000 ## number of permutations

sims={} ## store permutation results


def diversity_function(x,A,B): ## function to fit
    """
    compute cumulative diversity given discovery date x
    """
    return A*np.log2(1+ np.array(x)/B)


for sim in range(N): ## permute N times
    if sim%100==0:
        print(sim)
    
    sims[sim]={'sim': None, 'fit': None} ## will store permutted diversity under "sim" and predicted diversity under "fit"
    
    ys=[]
    done=set()
    
    all_tips=ll.getExternal() ## get all tips in tree
    
    shuffled_viruses=np.random.choice(all_tips,size=len(all_tips),replace=False) ## resample tips, i.e. will shuffle order

    for x in shuffled_viruses: ## iterate over tips in (now) random order
        cur_node=x ## begin at tip
        path_length=0.0 ## reset current path length to 0.0

        while cur_node and cur_node not in done: ## descend to root if node not seen previously
            path_length+=cur_node.length ## increment current path length
            done.add(cur_node) ## remember branch as done
            cur_node=cur_node.parent ## descend to parent

    #     ys.append(sum([k.length for k in done])/sum([k.length for k in ll.Objects]))
        ys.append(sum([k.length for k in done])) ## remember cumulative diversity at this point
    
    
    (A_, B_), _ = opt.curve_fit(diversity_function, list(xs), ys, maxfev=10000) ## fit curve to permutted data
    y_fit = diversity_function(list(xs), A_, B_) ## get expectation
    
    sims[sim]['sim']=ys ## store actual permuted diversity for later
    sims[sim]['fit']=y_fit ## store fitted diversity line for later

##########

print('done')
```
